<h1> Free vs Paid Apps </h1>
<img src="freevspaid.png">

```python
print("There are total of " + str(len(df.prime_genre.value_counts().index)) + " categories which is little too much")
print ("So we can limit our categories to 5")
```

```python
# reducing the number of categories

s = df.prime_genre.value_counts().index[:4]
def categ(x):
    if x in s:
        return x
    else : 
        return "Others"

df['broad_genre']= df.prime_genre.apply(lambda x : categ(x))
```

```python
free = df[df.price==0].broad_genre.value_counts().sort_index().to_frame()
paid = df[df.price>0].broad_genre.value_counts().sort_index().to_frame()
total = df.broad_genre.value_counts().sort_index().to_frame()
free.columns=['free']
paid.columns=['paid']
total.columns=['total']
dist = free.join(paid).join(total) # joining the individual columns
dist ['paid_per'] = dist.paid*100/dist.total
dist ['free_per'] = dist.free*100/dist.total
dist
```

```python
list_free= dist.free_per.tolist()
tuple_free = tuple(list_free)
tuple_paidapps = tuple(dist.paid_per.tolist())
```

```python
plt.figure(figsize=(15,8))
N=5
ind = np.arange(N)    # the x locations for the groups
width =0.56   # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, tuple_free, width, color='#45cea2')
p2 = plt.bar(ind, tuple_paidapps, width,bottom=tuple_free,color='#fdd470')
plt.xticks(ind,tuple(dist.index.tolist() ))
plt.legend((p1[0], p2[0]), ('free', 'paid'))
plt.show()
# for pie chart
pies = dist[['free_per','paid_per']]
pies.columns=['free %','paid %']
plt.show()
```

```python
plt.figure(figsize=(15,8))
pies.T.plot.pie(subplots=True,figsize=(20,4),colors=['#45cea2','#fdd470'])
plt.show()
```

### Insights
- Education has significant % of Paid apps.
- On the contrary - Entertainment category hosts high % of free apps

<h1> Hypotheses Testing </h1>

```python
import random
from scipy.stats import norm
from math import sqrt
col1=df["user_rating"]
list1=[]
i = 1
while(i<=1000):
    try:
        list1.append(df.at[random.choice(range(0,6000)),"user_rating"])
        i = i + 1
    except:
        continue
def one_sided_hypo(sample_mean, pop_mean, std_dev, sample_size, alpha):
    actual_z = abs(norm.ppf(alpha))
    hypo_z = (sample_mean - pop_mean) / (std_dev/sqrt(sample_size))
    print('actual z value :', actual_z)
    print('hypothesis z value :', hypo_z, '\n')
    if hypo_z >= actual_z:
        return True
    else:
        return False
    
alpha = 0.05
sample_mean = np.mean(list1)
pop_mean = np.mean(col1)
sample_size =  1000
std_dev = np.std(list1)

print("Sample mean=",sample_mean)
print("Standard Deviation=",std_dev)
print("Population mean=",pop_mean)
print("Sample Size=",sample_size)
print('H0 :  <=', pop_mean)
print('H1 :  >', pop_mean)
print('alpha value is :', alpha, '\n')

r = one_sided_hypo(sample_mean, pop_mean, std_dev, sample_size, alpha)
if r:
    print('Reject NULL hypothesis')
else:
    print('Failed to reject NULL hypothesis')
```
